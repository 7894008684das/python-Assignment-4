```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
```

```python
from sklearn import datasets
```

```python
a = datasets.load_iris()
y = a.target
iris_data = PCA(n_components = 3)
data = iris_data.fit_transform(a.data)
```

```python
y
```

```python
figure = plt.figure(1,figsize=(8,6))
ed = Axes3D(figure ,azim=-100,elev=10,)
ed.scatter(data[:,0],data[:,1],data[:,2],c=y)
ed.set_title('3D')
ed.set_xlabel('x_axis')
ed.set_ylabel('y_axis')
ed.set_zlabel('z_axis')

plt.show()
```
